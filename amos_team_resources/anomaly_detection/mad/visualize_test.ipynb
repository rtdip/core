```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import json
from pathlib import Path

# Add SDK to path
SDK_PATH = Path().resolve().parents[2] / "src" / "sdk" / "python"
sys.path.insert(0, str(SDK_PATH))

from rtdip_sdk.pipelines.visualization.plotly.anomaly_detection import AnomalyDetectionPlotly
from rtdip_sdk.pipelines.visualization.matplotlib.anomaly_detection import AnomalyDetectionPlot

from rtdip_sdk.pipelines.anomaly_detection.spark.mad.mad_anomaly_detection import MadAnomalyDetection

print("Imports complete")
```

```python
from pyspark.sql import SparkSession
from datetime import datetime, timedelta

session = SparkSession.builder.appName("PlotlyDemoAD").getOrCreate()


base = datetime(2024, 1, 1)

data = [
    (base + timedelta(seconds=i), v)
    for i, v in enumerate(
        [10.0, 12.0, 10.5, 11.0, 30.0, 10.2, 9.8, 10.1, 10.3, 10.0]
    )
]
columns = ["timestamp", "value"]

df = session.createDataFrame(data, columns)
```

```python
ad_scorer = MadAnomalyDetection()

anoms = ad_scorer.detect(df)
```

```python
plotly_visualizer = AnomalyDetectionPlotly(ts_data=df, ad_data=anoms, sensor_id="Plotly_Demo")
plotly_visualizer.plot()
plotly_visualizer.save("anomaly_detection_plot.html")
```
